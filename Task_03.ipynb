# TASK 03 : IMAGE TO PENCIL SKETCH WITH PYTHON

**Step 01 : Importing Libraries**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Step 02 : Creating image to pencil
sketch function**

```python
def image_to_pencil_sketch(image,pencil_thickness=21):
  img=cv2.imread(image)
  gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  inverted_gray_image=255-gray_image
  blurred_image=cv2.GaussianBlur(inverted_gray_image,(pencil_thickness,pencil_thickness),sigmaX=0,sigmaY=0)
  inverted_blurred_image=255-blurred_image
  pencil_sketch=cv2.divide(gray_image,inverted_blurred_image,scale=256.0)
  return pencil_sketch
```

**Step 03 : Importing image**

```python
image='dog_face.jpg'
pencil_thickness=35
pencil_sketch=image_to_pencil_sketch(image,pencil_thickness)
```

**Step 04 : Display the Original image and the Pencil Sketch**

```python
pencil_sketch=image_to_pencil_sketch(image)
fig,axes=plt.subplots(1,2,figsize=(20,10))
axes[0].imshow(cv2.cvtColor(cv2.imread(image),cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(pencil_sketch,cmap='gray')
axes[1].set_title('Pencil Sketch')
axes[1].axis('off')
plt.show()
```
